```python
Pokemon = data.sort_values(['Type 1', 'Total'], ascending=[1,0])
```

```python
Pokemon.head()
```

```python
Pokemon.tail()
```

```python
Grass = Pokemon.loc[(Pokemon['Type 1'] == 'Grass')].head(10)
Fire = Pokemon.loc[(Pokemon['Type 1'] == 'Fire')].head(10)
Water = Pokemon.loc[(Pokemon['Type 1'] == 'Water')].head(10)
Bug = Pokemon.loc[(Pokemon['Type 1'] == 'Bug')].head(10)
Normal = Pokemon.loc[(Pokemon['Type 1'] == 'Normal')].head(10)
Poison = Pokemon.loc[(Pokemon['Type 1'] == 'Poison')].head(10)
Electric = Pokemon.loc[(Pokemon['Type 1'] == 'Electric')].head(10)
Ground = Pokemon.loc[(Pokemon['Type 1'] == 'Ground')].head(10)
Fairy = Pokemon.loc[(Pokemon['Type 1'] == 'Fairy')].head(10)
Fighting = Pokemon.loc[(Pokemon['Type 1'] == 'Fighting')].head(10)
Psychic = Pokemon.loc[(Pokemon['Type 1'] == 'Psychic')].head(10)
Rock = Pokemon.loc[(Pokemon['Type 1'] == 'Rock')].head(10)
Ghost = Pokemon.loc[(Pokemon['Type 1'] == 'Ghost')].head(10)
Ice = Pokemon.loc[(Pokemon['Type 1'] == 'Ice')].head(10)
Dragon = Pokemon.loc[(Pokemon['Type 1'] == 'Dragon')].head(10)
Dark = Pokemon.loc[(Pokemon['Type 1'] == 'Dark')].head(10)
Steel = Pokemon.loc[(Pokemon['Type 1'] == 'Steel')].head(10)
Flying = Pokemon.loc[(Pokemon['Type 1'] == 'Flying')].head(10)
```

```python
All_top_ten_pokemon = pd.concat([Grass,Fire,Water,Bug,Normal,Poison,Electric,Ground,Fairy,Psychic,Rock,
                                Ghost,Ice,Dragon,Dark,Steel,Flying])
```

```python
All_top_ten_pokemon
```

```python
plt.figure(figsize=(20,10))
plt.bar(All_top_ten_pokemon["Type 1"],All_top_ten_pokemon["Total"])
plt.title('Overall best by types')
plt.xlabel('Total')
plt.ylabel('Type')

plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(All_top_ten_pokemon["Type 1"],All_top_ten_pokemon["Attack"])
plt.title('Top Attacking Pokemons by Type')
plt.xlabel('Type')
plt.ylabel('Attack')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(All_top_ten_pokemon["Type 1"],All_top_ten_pokemon["Defense"])
plt.title('Top Defensive Pokemons by Type')
plt.xlabel('Type')
plt.ylabel('Defense')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(All_top_ten_pokemon["Type 1"],All_top_ten_pokemon["Speed"])
plt.title('Top Fast Pokemons by Type')
plt.xlabel('Type')
plt.ylabel('Speed')
plt.show()
```
